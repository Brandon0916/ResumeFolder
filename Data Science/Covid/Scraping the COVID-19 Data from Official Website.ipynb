# COVID-19 Data Scraping

This is my study file for scraping the data from the official website.  

```python
#First import request to the website for the BC CDC
from urllib.request import urlopen

html=urlopen("http://www.bccdc.ca/health-info/diseases-conditions/covid-19/case-counts-press-statements").read().decode('utf-8')
#print(html)
```

```python
import re

res=re.findall(r"<title>(.+?)</title>",html,flags=re.DOTALL)
print("\nPage title is: ", res[0])
```

```python
#now we use the beautifulsoup
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,features="lxml")
print(soup.title)
print('\n',soup.p) 
```

```python
#find the csv file of the website 
csv_links=soup.find_all("a",{"href":re.compile('.*?\.csv')})

temp=[]
for link in csv_links:
    temp.append(link['href'])
    
    
#csv_links=link['href']
#csv_links
temp[0]
```

```python
data_button=soup.find_all("a",{"class":"phsa-rteElement-Button"})
data_button
for d in data_button:
   print(d.get_text())
```
